```python

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

## making df

```python
b5=pd.read_excel("/content/drive/MyDrive/big5.xlsx")
b5
```

```python
ocean = b5.iloc[0:5]

ocean
```

```python
values = b5.drop([0, 1, 2, 3, 4])
values = values.reset_index(drop=True)

values
```

## multiple star diagram

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
# original df : b5
# big5 df : ocean
# the other df : values
```

### ocean

```python
## ocean multiple star diagram

fig = go.Figure()

# Define the colors for each column
colors = {'지성': 'red', '한나': 'blue', '은하': 'yellow', '소희': 'green'}


# Iterate over each columns
for column in ['지성', '한나', '은하', '소희']:
    fig.add_trace(go.Scatterpolar(
        r=ocean[column],
        theta=ocean['Unnamed: 0'],
        fill='toself',
        name=column,
        line=dict(color=colors[column])
    ))

fig.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 5]
        )),
    showlegend=True
)

fig.show()
```

```python
## change the opacity

fig = go.Figure()

# Define the colors for each column
colors = {'지성': 'red', '한나': 'blue', '은하': 'yellow', '소희': 'green'}


# Iterate over each columns
for column in ['지성', '한나', '은하', '소희']:
    fig.add_trace(go.Scatterpolar(
        r=ocean[column],
        theta=ocean['Unnamed: 0'],
        fill='toself',
        name=column,
        fillcolor=colors[column],
        opacity=0.3
    ))

fig.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 5]
        )),
    showlegend=True
)

fig.show()
```

### values

```python
## multiple star diagram

fig = go.Figure()

# Define the colors for each column
colors = {'지성': 'red', '한나': 'blue', '은하': 'yellow', '소희': 'green'}


# Iterate over each columns
for column in ['지성', '한나', '은하', '소희']:
    fig.add_trace(go.Scatterpolar(
        r=values[column],
        theta=values['Unnamed: 0'],
        fill='toself',
        name=column,
        line=dict(color=colors[column])
    ))

fig.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 100]
        )),
    showlegend=True
)

fig.show()
```

## scatter plot

```python
# Define the colors for each column
colors = {'지성': 'red', '한나': 'blue', '은하': 'yellow', '소희': 'green'}
transparency = {'지성': 0.4, '한나': 0.4, '은하': 0.5, '소희': 0.4}

# Create a scatter plot for each column
fig = go.Figure()

for column in ['지성', '한나', '은하', '소희']:
    fig.add_trace(go.Scatter(
        x=values[column],
        y=values['Unnamed: 0'],
        mode='markers',
        name=column,
        marker=dict(color=colors[column], size=10, opacity=transparency[column])  # Set the color and size for each column
    ))

fig.update_layout(
    xaxis_title='Value',
    yaxis_title='Category',
    title='Scatter Plot for OCEAN',
    showlegend=True
)

fig.show()
```
