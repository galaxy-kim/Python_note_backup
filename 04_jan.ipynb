```python
#등분산 검정
#pvalue=0.6201403975363267 > 0.05 ← 귀무가설 채택

pd.options.display.float_format='{:.3f}'.format

type1=[]
for i in range(1,8,1):
    type1.append(df_sh[df_sh.Frequency==i].총_할인_금액)
sp.stats.levene(type1[0],type1[1],type1[2],type1[3],type1[4],type1[5],type1[6])
```

```python
#ANOVA 검정(등분산이다)

model=ols("총_할인_금액~Frequency",data=df_sh).fit()
anova_lm(model)
```
