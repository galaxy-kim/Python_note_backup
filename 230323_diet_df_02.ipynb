```python
from sklearn.feature_selection import SelectKBest,chi2
```

```python
from sklearn.model_selection import train_test_split
```

```python
age_4.info()
```

```python
#데이터 지정(변수)
X=age_4.drop(['id', 'created_YMD', 'order_id', 'p_price_000', 'p_price_000_idcount'],axis=1)
Y=age_4['p_price_000']

#데이터 분할
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)

#일변량 통계기반 변수 선택
selector=SelectKBest()
selector.fit(X_train,Y_train)

score=pd.DataFrame()
score["Chi2"]=selector.scores_
score["p-value"]=selector.pvalues_
score["support"]=selector.get_support()
score["attribute"]=X_train.columns
score[score["support"]==True]
```

### p_price_000에 대한 correlation coefficient

```python
# Calculate the correlation coefficient between each feature and 'p_price_000'
corr_coefficients = age_4[['occupation', 'sex', 'married', 'dise_yn', 'diet_purpose', 'p_category', 'p_month', 'bmi_g', 'p_price_000']].corr()['p_price_000'][:-1]

# Print the correlation coefficients in descending order
print(corr_coefficients.abs().sort_values(ascending=False))
```

```python
# Calculate the correlation matrix
corr_matrix = age_4[['occupation', 'sex', 'married', 'dise_yn', 'diet_purpose', 'p_category', 'p_month', 'bmi_g', 'p_price_000']].corr()

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True,fmt="d", cmap='coolwarm')
plt.show()
```
