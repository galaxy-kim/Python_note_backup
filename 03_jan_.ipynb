```python
from scipy import stats
```

```python
#일표본 t-test(표본 하나로 모수를 추정)

print(df2.총_매출액.mean())

print(stats.ttest_1samp(df2["총_매출액"],7700000))
```

```python
#평균을 그대로 넣는다면

print(df2.총_매출액.mean())

print(stats.ttest_1samp(df2["총_매출액"],7068689))
```

```python
#독립표본 t-test(가장 일반적인 경우)

no_claim=df2[df2["클레임접수여부"]==0]
no_claim
claim=df2[df2["클레임접수여부"]==1]
claim
```

```python
no_claim.방문빈도.mean()
```

```python
claim.방문빈도.mean()
```

```python
print(stats.ttest_ind(no_claim["방문빈도"],claim["방문빈도"],equal_var=False))

#등분산성 확인을 못했으니 equal_var=False 입력(등분산 검정 안해도 t-test 할 수 있다)
#등분산 검정을 먼저 할 경우 True로 넣어주면 됨
```

```python
#ANOVA

import pingouin as pg
import scikit_posthocs
import scipy as sp
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
df2["구매유형"].describe()
```

```python
df2.groupby("구매유형").mean()
```

```python
#등분산 분석 : 등분산이다(귀무), 등분산 아니다(대립) ← pvalue<0.05 대립가설 채택

df1=df2[["구매유형","총_매출액"]]
pd.options.display.float_format='{:.3f}'.format

type1=[]
for i in range(1,5,1):
    type1.append(df1[df1.구매유형==i].총_매출액)
sp.stats.levene(type1[0],type1[1],type1[2],type1[3])
```

```python
#등분산이 아닌 경우
#H0 각 구매유형 별로 총 매출액 평균이 같다
#H1 각 구매유형 별로 총 매출액 평균이 다르다

pg.welch_anova(dv="총_매출액",between="구매유형",data=df2)
```

```python
#등분산인 경우

model=ols("총_매출액~구매유형",data=df1).fit()
anova_lm(model)
```
