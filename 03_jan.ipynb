```python
q1=df_salary["salary"].quantile(q=0.25)
q3=df_salary["salary"].quantile(q=0.75)
IQR=q3-q1
IQR

#q1:1분위수
```

```python
data_IQR=df_salary[(df_salary["salary"]<q3+IQR*1.5)&(df_salary["salary"]>q1-IQR*1.5)]
data_IQR

#(이상치를 제거한)박스플롯 안 데이터만 뽑기
#자주 쓰이기 때문에 이 코드는 외워두는 것이 좋다(원데이터를 남겨두면서 정리된 데이터 사용가능)
```

```python
data_IQR["salary"].hist()

#정규분포 형태에 더 가까워짐
```

```python
under_500=df[df["고객ID"]<=500]
upper_500=df[df["고객ID"]>500]

print(under_500)
print(upper_500)

#행 나누기
```

```python
df_join=under_500.append(upper_500)
print(df_join)
```

```python
df_join.describe()
```

```python
y=df_join[["고객ID","총매출액"]]

#필요한 열만 골라서 가져오기
```

```python
x=df_join[["고객ID","방문빈도","고객등급","구매유형","거주지역"]]

#[인덱스[리스트]]
```

```python
y=pd.DataFrame(y)
x=pd.DataFrame(x)

#Series → DataFrame(단순한 배열을 엑셀화)
```

```python
xy_merge=y.merge(x)
xy_merge

#공통된 key에 맞춰 merge됨
```

```python
xy_merge.plot.scatter(x="방문빈도",y="총매출액",title="방문빈도와 총매출액의 관계")
#한글이라 깨져서 나옴
```
