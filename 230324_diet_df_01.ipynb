```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("/content/drive/MyDrive/Colab Notebooks/ubion_given_data/online_diet_age_4.xlsx")
df
```

```python
df=df.drop(['Unnamed: 0', 'age_code', 'created_YMD', 'order_id'],axis=1)
```

```python
# convert column B to numeric and replace non-numeric values with NaN
df['waist'] = pd.to_numeric(df['waist'], errors='coerce')
df['heap'] = pd.to_numeric(df['heap'], errors='coerce')

# replace NaN values with the mean value of column B
mean_B = np.mean(df[['waist', 'heap']])
df[['waist', 'heap']] = df[['waist', 'heap']].fillna(mean_B)
```

```python
df.info()
```

```python
cg=df[['id','occupation', 'sex', 'married', 'dise_yn', 'diet_purpose','pfid', 'p_category', 'p_month', 'bmi_g']]
num=df[['height', 'weight', 'want_diet', 'weight_reduce', 'waist', 'heap', 'p_price_000','p_price_000_idcount']]
```

```python
ndf=cg.join(num)
```

```python

# define the bins and labels for grouping
bins = [0, 1, 4, 9, float('inf')]
labels = ['1', '2', '3', '4']
## 1 : 1회구매, 2 : 2~4회구매, 3 : 5~9회구매, 4 : 10회이상

# create a new column with the groups based on 'Purchasing Frequency'
ndf['idcount_code'] = pd.cut(ndf['p_price_000_idcount'], bins=bins, labels=labels)

# print the resulting dataframe
ndf
```

```python
ndf.info()
```

## 검정

```python
import scipy as sp
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
     
```

### 등분산 "p_category", "p_month"

```python
#등분산test(1)
#귀무가설: 등분산이다.
#대립가설: 등분산이 아니다.

ndf_anova = ndf[["p_category", "p_month"]]
pd.options.display.float_format = '{:.4f}'.format #결과값 소숫점을 4째자리까지

type1 = []
for i in range(1,5,1): #4개의 집단 평균비교(#1,2,3,4)
    type1.append(ndf_anova[ndf_anova.p_category==i].p_month)
sp.stats.levene(type1[0],type1[1],type1[2],type1[3])

#결과값
#pvalue =  (<0.05)
#대립가설 채택 -> '등분산 아니다'
```

```python
# Import necessary libraries
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

### 챗gpt 모델

```python

# Load the data
data = ndf

# Prepare the data
X = data[['occupation', 'sex', 'married', 'dise_yn', 'diet_purpose', 'p_month', 'bmi_g', 'height', 'weight', 'want_diet', 'p_price_000']]
y = data['p_category']

# Encode categorical variables
X = pd.get_dummies(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create and fit the logistic regression model
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)

# Make predictions on the test set
y_pred = logreg.predict(X_test)

# Evaluate the model
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
print('Accuracy:', accuracy)
print('Confusion matrix:', confusion)
```

```python
# Get the coefficients and corresponding feature names
coef = logreg.coef_[0]
features = X.columns

# Create a DataFrame of coefficients and features
coef_df = pd.DataFrame({'feature': features, 'coefficient': coef})

# Sort the coefficients by absolute value in descending order
coef_df['abs_coef'] = abs(coef_df['coefficient'])
sorted_coef = coef_df.sort_values('abs_coef', ascending=False)

# Print the sorted coefficients
print(sorted_coef)
```
