```python
TM_txt=open("/content/drive/MyDrive/Colab Notebooks/ubion_given_data/textmining.txt","rt",encoding="UTF-8").read()
```

```python
TM_txt
```

```python
from konlpy.tag import Okt
from collections import Counter
```

```python
okt=Okt() #태그를 붙이는 함수
line=[]

line=okt.pos(TM_txt) #품사를 붙여주는 함수
n_adj=[]
for word, tag in line:
    if tag in ["Noun","Adjective"]: #이런 태그가 붙은 경우
      n_adj.append(word) #이 리스트 안에 단어를 넣어라
print(n_adj)
```

```python
#필요없는 단어 빼기
stop_words="마 이닝 등 수 및 있다 있는 이 빈 간의 통해 키 워드 대한 로 를 각"
stop_words=set(stop_words.split(" "))
n_adj=[word for word in n_adj if not word in stop_words]
```

```python
#몇번 등장했는지 카운팅
counts=Counter(n_adj)
tags=counts.most_common(20)
print(counts)
print(tags)
```

```python
from wordcloud import WordCloud
```

```python
import numpy as np
from PIL import Image
```

```python
font="/content/drive/MyDrive/Colab Notebooks/ubion_given_data/NanumSquareNeo-cBd.ttf"

word_cloud=WordCloud(font_path=font,background_color="white",
                     max_font_size=300,colormap="BuGn").generate_from_frequencies(dict(tags))
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.imshow(word_cloud)
plt.axis("off")
```
